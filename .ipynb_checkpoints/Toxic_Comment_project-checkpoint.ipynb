# Data Preprocessing

# Cleaning the data - Applying Stemming and Lemantizer

```python
wordnet=WordNetLemmatizer()
stemmer=PorterStemmer()
```

```python
def cleanComment(comment):
    comment=comment.lower()
    #remove \n
    comment=re.sub("\\n","",comment)
    words=tokenizer.tokenize(comment)
    words=[wordnet.lemmatize(word, "v") for word in words]
    words = [w for w in words if not w in stop]
    clean_sent=" ".join(words)
    return cleanComment
    
    
```

# Clean Corpus

```python
clean_corpus=train['comment_text'].apply(lambda x :cleanComment(x))
```

# TF-IDF Vectorizer

```python
vectorizer = TfidfVectorizer(strip_accents='unicode', analyzer='word', ngram_range=(1,3), norm='l2')
vectorizer.fit(train['comment_text'])
vectorizer.fit(test['comment_text'])
x_train = vectorizer.transform(train['comment_text'])
y_train = train.drop(labels = ['id','comment_text'], axis=1)
x_test = vectorizer.transform(valid['comment_text'])
y_test = valid.drop(labels = ['id','comment_text'], axis=1)
x_test_final = vectorizer.transform(test['comment_text'])
y_test_final = test.drop(labels = ['id','comment_text'], axis=1)
```

Converting train , valid and test into vectors using TF-IDF vectorization technique. 'x_test' - validation data. 'x_test_final' - Test data.

# Logistic Regression

```python
# Using pipeline for applying logistic regression and one vs rest classifier
C=[0.01,0.1,1,2,3]
regular=['l1','l2']
result_lr=np.zeros((len(regular),len(C),len(CATEGORIES)))
l_loss_lr_va=np.zeros((len(regular),len(C),len(CATEGORIES)))
l_loss_lr_tr=np.zeros((len(regular),len(C),len(CATEGORIES)))
pred_lr=np.zeros((len(regular),len(C),len(CATEGORIES)))
for j,reg in enumerate(regular):
    print("---",reg,"---")
    for k,c in enumerate(C):
        """LogReg_pipeline = Pipeline([
                        ('clf', OneVsRestClassifier(LogisticRegression(penalty=reg,C=c), n_jobs=-1)),
                    ])"""
        i=0
        l_loss=np.zeros(len(C))
        for i,category in enumerate(CATEGORIES):
            LogReg_pipeline = Pipeline([
                        ('clf', OneVsRestClassifier(LogisticRegression(penalty=reg,C=c), n_jobs=-1)),
                    ])
            print('**Processing {} comments...**'.format(category))

            # Training logistic regression model on train data
            LogReg_pipeline.fit(x_train,y_train[category])

            # calculating test accuracy
            prediction = LogReg_pipeline.predict(x_test)
            prediction_tr = LogReg_pipeline.predict(x_train)
            pred=LogReg_pipeline.predict_proba(x_test)[:,1]
            result_lr[j][k][i]=(roc_auc_score(y_test[category],pred))
            l_loss_lr_va[j][k][i]=(log_loss(y_test[category], prediction))
            l_loss_lr_tr[j][k][i]=(log_loss(y_train[category], prediction_tr))
            #fpr, tpr,_=roc_curve(pred[j][k][i],y_test[category],drop_intermediate=False)
            #ax[k+j][i].plot(fpr, tpr, color='red',lw=2, label='ROC curve')
            print('AUC score is {} for C={} and Reg={}'.format(result_lr[j][k][i],c,reg))
            print("\n")
```

# Evaluation for Logistic Regression using TEST DataSet

```python
"""LogReg_pipeline = Pipeline([
                        ('clf', OneVsRestClassifier(LogisticRegression(penalty='l1',C=1), n_jobs=-1)),
                    ])"""
auc=np.zeros(len(CATEGORIES))
for i,category in enumerate(CATEGORIES):
    LogReg_pipeline = Pipeline([
                        ('clf', OneVsRestClassifier(LogisticRegression(penalty='l1',C=1), n_jobs=-1)),
                    ])
    LogReg_pipeline.fit(x_train,y_train[category])
    #prediction = LogReg_pipeline.predict(x_test)
    pred=LogReg_pipeline.predict_proba(x_test_final)[:,1]
    fpr, tpr,_=roc_curve(y_test_final[category],pred,drop_intermediate=False)
    plt.plot(fpr, tpr,lw=2, label='ROC curve')
    plt.xlabel("True Positive Rate")
    plt.ylabel("False Positive Rate")
    plt.title("Penalty = l1 C=1 - Logistic Regression ROC curve")
    print("For category -", category, roc_auc_score(y_test_final[category],pred))
    auc[i]=(roc_auc_score(y_test_final[category],pred))
plt.legend(CATEGORIES)
```

```python
print("Mean AUC for Logistic Regression with l1 regularisation and C=1 ----", np.mean(auc))
```

# LSTM - KERAS

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.layers import Dense, Input, LSTM, Embedding, Dropout, Activation
from keras.layers import Bidirectional, GlobalMaxPool1D
from keras.models import Model
from keras import initializers, regularizers, constraints, optimizers, layers
```

```python
list_classes = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
y = train[list_classes].values
list_sentences_train = train["comment_text"]
list_sentences_test = test["comment_text"]
max_features = 20000
tokenizer = Tokenizer(num_words=max_features)
tokenizer.fit_on_texts(list(list_sentences_train))
list_tokenized_train = tokenizer.texts_to_sequences(list_sentences_train)
list_tokenized_test = tokenizer.texts_to_sequences(list_sentences_test)
```

```python
maxlen = 200
X_t = pad_sequences(list_tokenized_train, maxlen=maxlen)
X_te = pad_sequences(list_tokenized_test, maxlen=maxlen)
```

```python
inp = Input(shape=(maxlen, ))
```

```python
embed_size = 128
x = Embedding(max_features, embed_size)(inp)
```

```python
x = LSTM(60, return_sequences=True,name='lstm_layer')(x)
```

```python
x = GlobalMaxPool1D()(x)
```

```python
x = Dropout(0.1)(x)
```

```python
x = Dense(50, activation="relu")(x)
```

```python
x = Dropout(0.1)(x)
```

```python
x = Dense(6, activation="sigmoid")(x)
```

```python
model = Model(inputs=inp, outputs=x)
model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
```

```python
batch_size = 32
epochs = 2
model.fit(X_t,y, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

# HyperParameter Tuning - Loss vs epochs

```python
epochs = [2,4,5]
loss=[]
val_loss=[]
for i in epochs:
    x = Embedding(max_features, embed_size)(inp)
    x = LSTM(60, return_sequences=True,name='lstm_layer')(x)
    x = GlobalMaxPool1D()(x)
    x = Dropout(0.1)(x)
    x = Dense(50, activation="relu")(x)
    x = Dropout(0.1)(x)
    x = Dense(6, activation="sigmoid")(x)
    model = Model(inputs=inp, outputs=x)
    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    batch_size = 32
    history=model.fit(X_t,y, batch_size=batch_size, epochs=i, validation_split=0.33)
    loss.append(history.history['loss'])
    val_loss.append(history.history['val_loss'])
```

```python
print(loss)
plot_loss=[]
plot_val_loss=[]
for i in loss:
    plot_loss.append(i[len(i)-1])
for i in val_loss:
    plot_val_loss.append(i[len(i)-1])
print(plot_loss)
```

```python
plt.plot(epochs,plot_loss,color="blue",label="training")
plt.plot(epochs,plot_val_loss,color="red",label="validation")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("loss")
plt.title("Model Used - LSTM - Loss vs Epochs")
plt.show()
```
